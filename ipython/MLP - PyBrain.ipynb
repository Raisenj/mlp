### Building the Dataset

We will create a Dataset using the SupervisedDataSet module from pybrain.datasets

```python
from pybrain.datasets import SupervisedDataSet
ds = SupervisedDataSet(20,1)
```

Now we need to read the dataset mentioned above and insert the items.

```python
f = open('../dataset/german.numeric.txt', 'r')
for line in f:
    # We extract the items from line
    items = line.split()
    # We store in target the solution
    target = items[len(items)-1]
    # We remove the last item that corresponds to the target
    items = items[:len(items)-1]
    # Now we have in items the attributes and in target the solution
    # Hence, we are going to add the sample to the dataset
    ds.addSample(items, target)
```

### Splitting the DataSet

In order to perform train, validation and test we need to split the current dataset. Hence, we will use 70% for train, 15% for validation and 15% for testing.

## BUILDING THE NETWORK - MLP

We use the *buildNetwork* Shortcut of **PyBrain** to build the MLP. We will train different models (5,10,15 and 20 neurons in the hidden layer) and with the validation dataset we will choose the best one. Then we will test the best model with the testing set.

```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.structure import TanhLayer, SoftmaxLayer
net1 = buildNetwork(20,5,1, hiddenclass=TanhLayer, bias=True)
net1.randomize()
net2 = buildNetwork(20,10,1, hiddenclass=TanhLayer, bias=True)
net2.randomize()
net3 = buildNetwork(20,15,1, hiddenclass=TanhLayer, bias=True)
net3.randomize()
net4 = buildNetwork(20,20,1, hiddenclass=TanhLayer, bias=True)
net4.randomize()
```

We can run the network given an instance by using the *activate* function

```python
net1.activate([2, 10, 5, 7, 2241, 3, 4, 1, 3, 3, 3, 1, 48, 2, 1, 2, 3, 2, 1, 2])
```

## Training our Dataset

We will train our dataset with the BackpropTrainer from **PyBrain** which executes the Backpropagation approach.

```python
from pybrain.supervised.trainers import BackpropTrainer
trainer1 = BackpropTrainer(net1, nds, learningrate=0.1, momentum=0.6, batchlearning=False)
trainer2 = BackpropTrainer(net2, nds, learningrate=0.1, momentum=0.6, batchlearning=False)
trainer3 = BackpropTrainer(net3, nds, learningrate=0.1, momentum=0.6, batchlearning=False)
trainer4 = BackpropTrainer(net4, nds, learningrate=0.1, momentum=0.6, batchlearning=False)
```

We have now the trainer. We can execute manually one epoch by:

```python
trainer1.train()
```

Or execute it until it converges:

```python
errors1 = trainer1.trainUntilConvergence(maxEpochs=1000, verbose=True,validationProportion=0.2)
```

```python
errors2 = trainer2.trainUntilConvergence(maxEpochs=1000, verbose=True,validationProportion=0.2)
```

```python
errors3 = trainer3.trainUntilConvergence(maxEpochs=1000, verbose=True, validationProportion=0.2)
```

```python
error4 = trainer4.trainUntilConvergence(maxEpochs=1000, verbose=True,validationProportion=0.2)
```
