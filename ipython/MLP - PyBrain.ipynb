## Building the Dataset

We will create a Dataset using the SupervisedDataSet module from pybrain.datasets

```python
from pybrain.datasets import SupervisedDataSet
ds = SupervisedDataSet(20,1)
```

Now we need to read the dataset mentioned above and insert the items.

```python
f = open('../dataset/german.numeric.txt', 'r')
for line in f:
    # We extract the items from line
    items = line.split()
    # We store in target the solution
    target = items[len(items)-1]
    # We remove the last item that corresponds to the target
    items = items[:len(items)-1]
    # Now we have in items the attributes and in target the solution
    # Hence, we are going to add the sample to the dataset
    ds.addSample(items, target)
```

We are going to examine a little the dataset.

## Building a Network

First we are going to use the *buildNetwork* Shortcut of **PyBrain**

```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.structure import TanhLayer
net = buildNetwork(20,10,1, hiddenclass=TanhLayer, bias=True)
```

In the above code we built a network with 20 inputs, 10 hidden neurons in the hidden layer and a single output layer

The net is already initialized. We can use it with the activate function. For example:

```python
net.activate([2, 10, 5, 7, 2241, 3, 4, 1, 3, 3, 3, 1, 48, 2, 1, 2, 3, 2, 1, 2])
```

We can also examine a little bit the structure of the network

```python
net['in']
```

```python
net['hidden0']
```

```python
net['out']
```

Hence, we already build the network!!!

## Training our Dataset

We will train our dataset with the BackpropTrainer from **PyBrain**

```python
from pybrain.supervised.trainers import BackpropTrainer
trainer = BackpropTrainer(net, ds)
```

We have now the trainer. We can execute manually one epoch by:

```python
trainer.train()
```

Or execute it until it converges:

```python
trainer.trainUntilConvergence(maxEpochs=1000,verbose=True)
```
